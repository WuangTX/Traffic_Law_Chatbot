1. Import thư viện

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from pathlib import Path
import pandas as pd
```

2. Converting text using TF-IDF

```python
def find_project_root(start_path: Path) -> Path:
    current = start_path.resolve()
    for p in [current] + list(current.parents):
        if (p / "data").exists():
            return p
    return current


project_root = find_project_root(Path.cwd())
corpus_path = project_root / "data" / "processed" / "corpus_raw.txt"

if not corpus_path.exists():
    raise FileNotFoundError(
        f"Khong tim thay file corpus: {corpus_path}."
    )


def corpus_generator(file_path: Path):
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                yield line


print(f"Bat dau xu ly file: {corpus_path.name}")
print("Dang khoi tao ...")

vectorizer = TfidfVectorizer()

print("Dang ma hoa du lieu...")
corpus_list = list(corpus_generator(corpus_path))
X_encoded = vectorizer.fit_transform(corpus_list) 

print("\n--- KET QUA ---")
df = pd.DataFrame(X_encoded.toarray(), columns=vectorizer.get_feature_names_out(), index = corpus_list)
df.head() 
```
