1. Import thư viện 

```python
from pathlib import Path
import scipy.sparse
from sklearn.feature_extraction.text import HashingVectorizer
```

```python
import json
from pathlib import Path


def normalize_text(value: str) -> str:
    """Chuan hoa xuong dong va khoang trang trong van ban."""
    value = value.replace("\r\n", " ").replace("\n", " ").replace("\r", " ")
    value = value.replace("\\n", " ")
    return " ".join(value.split())


def extract_token_corpus(obj, keys=("content_tokenized", "content_token", "text_tokenized")):
    """Lay corpus tu cac truong da tokenization trong JSON."""
    texts = []

    if isinstance(obj, dict):
        for k, v in obj.items():
            if k in keys and isinstance(v, str):
                clean_text = normalize_text(v)
                if clean_text:
                    texts.append(clean_text)
            texts.extend(extract_token_corpus(v, keys=keys))
    elif isinstance(obj, list):
        for item in obj:
            texts.extend(extract_token_corpus(item, keys=keys))

    return texts


# Tim root cua project de doc file du lieu on dinh
def find_project_root(start_path: Path) -> Path:
    current = start_path.resolve()
    for p in [current] + list(current.parents):
        if (p / "data").exists() and (p / "encoding").exists():
            return p
    return current

project_root = find_project_root(Path.cwd())

# Uu tien file sau tokenization tu notebook pre_processing
candidate_jsons = [
    project_root / "data" / "processed" / "preprocessed_data_1.json",
]

json_path = next((p for p in candidate_jsons if p.exists()), candidate_jsons[0])
output_path = project_root / "data" / "processed" / "corpus_raw.txt"

with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

corpus = extract_token_corpus(data)

# Loai bo ban ghi trung lap, giu nguyen thu tu
corpus = list(dict.fromkeys(corpus))

with open(output_path, "w", encoding="utf-8") as f:
    for line in corpus:
        f.write(line + "\n")

print(f"Doc JSON: {json_path}")
print(f"Da tao corpus: {output_path}")
print(f"So doan van ban tokenized: {len(corpus)}")
print("\nMau 3 doan dau:")
for i, sample in enumerate(corpus[:3], 1):
    print(f"{i}. {sample[:200]}...")
```

3. Hash_vectorizing

```python
def find_project_root(start_path: Path) -> Path:
    current = start_path.resolve()
    for p in [current] + list(current.parents):
        if (p / "data").exists():
            return p
    return current


project_root = find_project_root(Path.cwd())
corpus_path = project_root / "data" / "processed" / "corpus_raw.txt"
output_matrix_path = project_root / "data" / "processed" / "hashed_matrix.npz"

if not corpus_path.exists():
    raise FileNotFoundError(
        f"Khong tim thay file corpus: {corpus_path}. Hay chay Cell 3 truoc de tao corpus_raw.txt"
    )


# Ham generator: doc tung dong de tiet kiem RAM
def corpus_generator(file_path: Path):
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                yield line


print(f"Bat dau xu ly file: {corpus_path.name}")
print("Dang khoi tao Hashing Vectorizer...")

vectorizer = HashingVectorizer(
    n_features=2**14,
    alternate_sign=False,
    token_pattern=r"(?u)\b\w+\b"
 )

print("Dang ma hoa du lieu...")
X_encoded = vectorizer.transform(corpus_generator(corpus_path))

print("\n--- KET QUA ---")
print(f"Kich thuoc ma tran: {X_encoded.shape}")
print(f"So luong van ban (hang): {X_encoded.shape[0]}")
print(f"So chieu vector (cot): {X_encoded.shape[1]}")


scipy.sparse.save_npz(output_matrix_path, X_encoded)
print(f"Da luu file ma tran thanh cong tai: {output_matrix_path}")
```

```python
project_root = find_project_root(Path.cwd())
corpus_path_2 = project_root / "data" / "processed" / "corpus_raw_2.txt"
output_matrix_path_2 = project_root / "data" / "processed" / "hashed_matrix_2.npz"

if not corpus_path_2.exists():
    raise FileNotFoundError(
        f"Khong tim thay file corpus: {corpus_path_2}."
    )

print(f"Bat dau xu ly file: {corpus_path_2.name}")
print("Dang khoi tao Hashing Vectorizer...")

vectorizer = HashingVectorizer(
    n_features=2**14,
    alternate_sign=False,
    token_pattern=r"(?u)\b\w+\b"
 )

print("Dang ma hoa du lieu...")
X_encoded_2 = vectorizer.transform(corpus_generator(corpus_path_2))

print("\n--- KET QUA ---")
print(f"Kich thuoc ma tran: {X_encoded_2.shape}")
print(f"So luong van ban (hang): {X_encoded_2.shape[0]}")
print(f"So chieu vector (cot): {X_encoded_2.shape[1]}")


scipy.sparse.save_npz(output_matrix_path_2, X_encoded_2)
print(f"Da luu file ma tran thanh cong tai: {output_matrix_path_2}")
```
